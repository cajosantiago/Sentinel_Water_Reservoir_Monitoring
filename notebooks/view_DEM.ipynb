```python
import matplotlib.pyplot as plt
import numpy as np

DEM = np.load('/home/csantiago/generated_data/DEM/Maranhão/DEM.npy')
SAM = np.load('/home/csantiago/generated_data/DEM/Maranhão/SAM.npy')

#DEM[SAM==SAM.min()] = DEM.max()+1
im0 = plt.imshow(DEM, cmap="terrain", origin="upper")
plt.colorbar(im0, label="Elevation (m)", fraction=0.046)
plt.show()

im1 = plt.imshow(SAM, cmap="terrain", origin="upper")
plt.colorbar(im1, label="Accumulation", fraction=0.046)
plt.show()
```
